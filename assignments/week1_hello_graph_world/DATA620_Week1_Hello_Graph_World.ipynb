## Objective

The goal of this notebook is to create a simple graph using Python and NetworkX, visualize the graph, and briefly describe what the graph structure shows. For this first network analysis exercise, I use the Krackhardt Kite social network because it is a small, well-known graph commonly used to introduce nodes, edges, and basic centrality concepts.

```python
import warnings
warnings.filterwarnings("ignore")

import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd

print("NetworkX version:", nx.__version__)
print("Pandas version:", pd.__version__)
```

## Create the graph

A graph is made of nodes and edges. In this example, each node represents a person and each edge represents a relationship between two people. This assignment uses the Krackhardt Kite graph shown in the assignment prompt.

```python
G = nx.Graph()

nodes = [
    "Andre", "Beverly", "Carol", "Diane", "Ed",
    "Fernando", "Garth", "Heather", "Ike", "Jane"
]

edges = [
    ("Andre", "Beverly"),
    ("Andre", "Carol"),
    ("Andre", "Diane"),
    ("Andre", "Fernando"),
    ("Beverly", "Diane"),
    ("Beverly", "Ed"),
    ("Beverly", "Garth"),
    ("Carol", "Diane"),
    ("Carol", "Fernando"),
    ("Diane", "Ed"),
    ("Diane", "Fernando"),
    ("Diane", "Garth"),
    ("Ed", "Garth"),
    ("Fernando", "Garth"),
    ("Fernando", "Heather"),
    ("Garth", "Heather"),
    ("Heather", "Ike"),
    ("Ike", "Jane")
]

G.add_nodes_from(nodes)
G.add_edges_from(edges)

print("Number of nodes:", G.number_of_nodes())
print("Number of edges:", G.number_of_edges())
```

## Visualize the graph

The visualization below shows the relationship structure among the people in the Krackhardt Kite network.

```python
pos = {
    "Andre": (0, 2),
    "Beverly": (0, 0),
    "Carol": (1, 3),
    "Diane": (1, 1),
    "Ed": (1, -1),
    "Fernando": (2, 2),
    "Garth": (2, 0),
    "Heather": (3, 1),
    "Ike": (4, 1),
    "Jane": (5, 1)
}

plt.figure(figsize=(10, 6))

nx.draw_networkx_nodes(G, pos, node_size=1800)
nx.draw_networkx_edges(G, pos, width=2)
nx.draw_networkx_labels(G, pos, font_size=10, font_weight="bold")

plt.title("Krackhardt Kite Social Network")
plt.axis("off")
plt.tight_layout()
plt.show()
```

## Basic graph summary

The summary below shows basic information about the network, including the number of nodes, number of edges, density, and average degree.

```python
summary = {
    "Number of nodes": G.number_of_nodes(),
    "Number of edges": G.number_of_edges(),
    "Is connected": nx.is_connected(G),
    "Density": round(nx.density(G), 3),
    "Average degree": round(sum(dict(G.degree()).values()) / G.number_of_nodes(), 2)
}

summary_df = pd.DataFrame(summary.items(), columns=["Metric", "Value"])
summary_df
```

## Node degree table

Degree counts how many direct connections each person has. People with higher degree are more directly connected in the network.


```python
degree_table = pd.DataFrame({
    "Person": list(G.nodes()),
    "Degree": [G.degree(node) for node in G.nodes()],
    "Degree Centrality": [nx.degree_centrality(G)[node] for node in G.nodes()]
})

degree_table = degree_table.sort_values(
    by=["Degree", "Person"],
    ascending=[False, True]
).reset_index(drop=True)

degree_table["Degree Centrality"] = degree_table["Degree Centrality"].round(3)

degree_table
```

## Interpretation

The graph contains 10 people and 18 relationships. The network is connected, meaning every person can be reached from every other person through some path.

Diane has the highest degree with 6 direct connections, making Diane the most directly connected person in this network. Fernando and Garth are also highly connected, each with 5 direct connections. Andre and Beverly each have 4 direct connections, so they also play important roles in the structure of the network.

Heather is important because she connects the main cluster of the network to Ike and Jane. Even though Heather does not have the highest degree, Heather acts as a bridge between the central group and the tail of the graph.
