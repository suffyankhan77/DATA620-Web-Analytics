## Objective

The objective of this assignment is to create and visualize a basic graph using Python and NetworkX. For this introductory graph analysis exercise, we use the Krackhardt Kite social network, where each node represents a person and each edge represents a relationship between two people.

This notebook demonstrates the basic workflow of graph analysis: defining nodes, defining edges, creating a graph object, checking the number of nodes and edges, and visualizing the final network structure.

```python
import networkx as nx         # here we import the libraries needed for networks and graphcs
import matplotlib.pyplot as plt

G = nx.Graph()   # we start with an empty graph

nodes = [     # here we create the nodes, 10 in total
    "Andre", "Beverly", "Carol", "Diane", "Ed",
    "Fernando", "Garth", "Heather", "Ike", "Jane"
]

G.add_nodes_from(nodes)      # on this line of code we add the above nodes to G

edges = [   # here we make the connections to each node for a total of 18 connections
    ("Andre", "Beverly"),
    ("Andre", "Carol"),
    ("Andre", "Diane"),
    ("Andre", "Fernando"),
    ("Beverly", "Diane"),
    ("Beverly", "Ed"),
    ("Beverly", "Garth"),
    ("Carol", "Diane"),
    ("Carol", "Fernando"),
    ("Diane", "Ed"),
    ("Diane", "Fernando"),
    ("Diane", "Garth"),
    ("Ed", "Garth"),
    ("Fernando", "Garth"),
    ("Fernando", "Heather"),
    ("Garth", "Heather"),
    ("Heather", "Ike"),
    ("Ike", "Jane")
]

G.add_edges_from(edges)  # here we add all the connections to G


print("Number of nodes:", G.number_of_nodes())  # 10 nodes / names
print("Number of edges:", G.number_of_edges())  # 18 edges / connections / relationships
```

```python
plt.figure(figsize=(8,6))   # finally we plot the kite graph
nx.draw_networkx(
    G,
    node_color="lightblue",
    node_size=2000,
    font_size=10
)

plt.title("Krackhardt Kite Graph")
plt.axis("off")
plt.show()
```
